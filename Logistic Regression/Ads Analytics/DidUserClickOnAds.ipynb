```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
ad_data = pd.read_csv("advertising.csv")
ad_data.head()
```

```python
ad_data.describe()
```

```python
sns.set_style('whitegrid')
ad_data['Age'].hist(bins=30)
plt.xlabel('Age')
```

```python
sns.jointplot('Age','Area Income',data=ad_data)
```

```python
sns.jointplot('Daily Time Spent on Site','Age',data=ad_data)
```

```python
sns.jointplot(x='Age',y='Daily Time Spent on Site',data=ad_data,color='red',kind='kde')
```

```python
sns.jointplot(x='Daily Time Spent on Site',y='Daily Internet Usage',data=ad_data,color='brown')
```

```python
sns.pairplot(ad_data,hue='Clicked on Ad')
```
