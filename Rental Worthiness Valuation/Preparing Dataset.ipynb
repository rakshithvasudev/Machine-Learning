```python
import numpy as np
import pandas as pd
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
ficoscores = np.random.randint(350,850,500)
ficoscores
```

```python
length_of_employment = np.random.randint(low=0, high=60, size = 500)
length_of_employment
```

```python
length_of_current_residency = np.random.randint(low=0, high=36, size=500)
length_of_current_residency
```

```python
total_monthly_income = np.random.randint(2000,10000,500)
total_monthly_income
```

```python
debt_to_income_ratio = np.random.uniform(low=0.0, high=1, size=500)
debt_to_income_ratio = debt_to_income_ratio * 100
```

```python
main = pd.DataFrame(data=[ficoscores,length_of_employment,length_of_current_residency,total_monthly_income,debt_to_income_ratio]).T
main
```

```python
main.to_csv("dataset.csv")
```
